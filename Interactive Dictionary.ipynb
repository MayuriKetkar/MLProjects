```python
import json
from difflib import SequenceMatcher, get_close_matches
```

```python
with open("dictionary.json") as f:
    data = json.load(f)
```

```python
for item in data:
    print(item)
```

```python
def getMeaning(word):
    if word in data:
        return data[word]
    elif word.title() in data:
        return data[word.title()]
    elif word.upper() in data:
        return data[word.upper()]
    elif len(get_close_matches(word, data.keys())) > 0:
        action = input("Did you mean %s instead? [y or n]: " % get_close_matches(word, data.keys())[0])
        while(action!="q"):
            if(action=="y"):
                return data[get_close_matches(word, data.keys())[0]]
            elif (action=="n"):
                res=input("The word doesn't exist. Do you want to add in the dictionary? [y or n]:")
                if(res=="y"):
                    meaning=input("Enter the description of the word you enterted:")
                    # print("you have entered:",meaning)
                    data.update({word :meaning})
                    return data[word]
                else:
                    return
            elif (action=="q"):
                return "You have quit"
            else:
                print("We don't understand your entry. Please Enter y - yes, n-no, q-quit.")
                action=input()
       
        
```

```python
word_user=input("Enter a word to search:")
output=getMeaning(word_user)
if type(output) == list:
    for item in output:
        print("-",item)
#For words having single definition
else:
    print("-",output)
```
